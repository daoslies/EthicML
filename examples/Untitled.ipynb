```python
import pandas as pd
```

```python
no_transform = pd.read_csv(f"seed_{0}_no_transform_reduced.csv")
for i in range(1, 10):
    temp = pd.read_csv(f"seed_{i}_no_transform_reduced.csv")
    no_transform = pd.concat([no_transform, temp], axis=0)
no_transform = no_transform.reset_index()
```

```python
no_transform.groupby(["model"]).agg(["mean", "std"])[['Accuracy', 'TPR_sensitive_0-sensitive_1', 'TPR_sensitive_0/sensitive_1']].round(3)
```
